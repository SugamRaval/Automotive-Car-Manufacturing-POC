```python
from prettytable import PrettyTable
import random
import os

class Car:
    def __init__(self, model_code, model_type, engine_type, car_type, color,
                 engine_capacity=None, tank_capacity=None,
                 battery_capacity=None, num_doors=None, max_speed=None,
                 safety_features=None, sound_system=None):
        self.model_code = model_code
        self.model_type = model_type
        self.engine_capacity=engine_capacity
        self.engine_type = engine_type
        self.car_type = car_type
        self.color = color
        self.tank_capacity = tank_capacity
        self.battery_capacity = battery_capacity
        self.num_doors = num_doors
        self.max_speed = max_speed
        self.safety_features = safety_features
        self.sound_system = sound_system
    
    

#-------------------------------------------------------------------------------

class M01(Car):
    def __init__(self, car_type, color,engine_capacity, tank_capacity,
                 battery_capacity, num_doors, max_speed,safety_features,
                 sound_system):
        super().__init__("M01", "Manual", "Gasoline", car_type, color,
                         engine_capacity, tank_capacity, battery_capacity,
                         num_doors, max_speed, safety_features, sound_system)

class M02(Car):
    def __init__(self, car_type, color,engine_capacity, tank_capacity,
                 battery_capacity, num_doors, max_speed,safety_features,
                 sound_system):
        super().__init__("M02", "Automatic", "Gasoline", car_type, color,
                         engine_capacity, tank_capacity,battery_capacity,
                         num_doors, max_speed, safety_features, sound_system)

class M03(Car):
    def __init__(self, car_type, color,engine_capacity, tank_capacity,
                 battery_capacity, num_doors, max_speed,safety_features,
                 sound_system):
        super().__init__("M03", "Automatic", "Electric", car_type, color,
                         engine_capacity, tank_capacity,battery_capacity,
                         num_doors, max_speed, safety_features, sound_system)

class M04(Car):
    def __init__(self, car_type, color,engine_capacity, tank_capacity,
                 battery_capacity, num_doors,max_speed, safety_features,
                 sound_system):
        super().__init__("M04", "Automatic", "Gasoline and Electric", car_type,
                         color,engine_capacity,tank_capacity, battery_capacity,
                         num_doors, max_speed,safety_features, sound_system)

#-------------------------------------------------------------------------------

def get_user_input(prompt, error_message, input_type):
  while True:
    try:
      user_input = input(prompt)
      return input_type(user_input)
    except:
      print(error_message)

#-------------------------------------------------------------------------------

def manufacture_car(user_obj):
    print(f"\nWelcome {user_obj.user_name}( Register ID : {user_obj.user_reg_id})")
    print("Available Models:")
    print('''
    1: M01 - Manual and Gasoline
    2: M02 - Automatic and Gasoline
    3: M03 - Automatic and Electric
    4: M04 - Automatic and Gasoline & Electric both''')
    while True:
        try:
            model_code = get_user_input("Enter Model Code From Above: ",
                                        "Invalid model code. Please try again!",
                                        int)
            if model_code in range(1,5):
                break
            else:
              raise ValueError("Invalid model code")

        except ValueError as e:
            print("Error: ",e)

    color_list=["Black","Blue","Red","Gray","Green","Yellow","Purple","Pink",
                "White","Orange"]
    while True:
      try:
        color = get_user_input("Enter Color Of Car: ",
                               "Invalid color. Please try again!", str)
        if color.title() in color_list:
          break
        else:
          raise ValueError("Invalid Color. Enter valid color.")

      except ValueError as e:
        print("Error: ",e)

    car_type_list=["Sedan","SUV"]
    car_type_list_check=[i.lower() for i in car_type_list]
    while True:
      try:
        car_type = get_user_input("Enter Car Type (Sedan or SUV): ",
                                  "Invalid car type. Please try again!", str)
        if car_type.lower() in car_type_list_check :
          break
        else:
          raise ValueError("Invalid Input. Enter Valid Car Type.")

      except ValueError as e:
        print("Error: ",e)
#-------------
    if model_code == 1:
       engine_capacity = "1500 CC"
       tank_capacity = 40.0
       battery_capacity = None
    elif model_code == 2:
         engine_capacity = "1300 CC"
         tank_capacity = 40.0
         battery_capacity = None
    elif model_code == 3:
         engine_capacity = "1300 CC"
         tank_capacity = None
         battery_capacity = 50.0
    elif model_code == 4:
         engine_capacity = "1300 CC"
         tank_capacity = 20.0
         battery_capacity = 25.0
#---------------
    while True:
      try:
        num_doors = get_user_input("Enter Number Of Doors (2 or 4 doors): ",
                                   "Invalid No. of Doors. Pleasae Try Again!",
                                   int)
        if num_doors in [2,4]:
            break
        else:
          raise ValueError("Invalid Input.Please Try Again in (2 or 4). ")
      except ValueError as e:
        print("Error: ",e)

    while True:
      try:
        max_speed = get_user_input("Enter Maximum Speed (120-300 km/h): ",
                                   "Invalid Speed. Pleasae Try Again!", float)
        if max_speed >= 120 and max_speed <= 300:
            break
        else:
          raise ValueError("Invalid speed! Enter into (120-300).")
      except ValueError as e:
        print("Error: ",e)

    # safety_features = input("Do you want to include Safety Features?")=='yes' or 'Yes' or 'YES'

    while True:
      try:
        safety_features = get_user_input("Do you want to include Safety Features?(Yes/No): ",
                                         "Invalid Input. Enter Yes or No only.",str)
        if safety_features.isdigit():
          raise ValueError("Invalid Input. Enter Yes or No only.")
        if safety_features in ["yes","Yes","YES"]:
          safety_features="Included"
          break
        elif safety_features in ["no","No","NO"]:
          safety_features="Not Included"
          break
        else:
          raise ValueError("Invalid Input. Enter Yes or No only.")
      except ValueError as e:
        print("ERROR: ",e)

    # sound_system = input("Do you want to include Sound System ?").lower()=='yes' or 'Yes'or 'YES'
    while True:
      try:
        sound_system = get_user_input("Do you want to include Sound System?(Yes/No): ",
                                         "Invalid Input. Enter Yes or No only.",str)
        if sound_system.isdigit():
          raise ValueError("Invalid Input. Enter Yes or No only.")
        if sound_system in ["yes","Yes","YES"]:
          sound_system="Included"
          break
        elif sound_system in ["no","No","NO"]:
          sound_system="Not Included"
          break
        else:
          raise ValueError("Invalid Input. Enter Yes or No only.")
      except ValueError as e:
        print("ERROR: ",e)

    if model_code == 1:
        return M01(car_type, color, engine_capacity, tank_capacity,
                   battery_capacity, num_doors,max_speed, safety_features,
                   sound_system)
    elif model_code == 2:
       return M02(car_type, color, engine_capacity, tank_capacity,
                  battery_capacity, num_doors,max_speed, safety_features,
                  sound_system)
    elif model_code == 3:
       return M03(car_type, color, engine_capacity, tank_capacity,
                  battery_capacity, num_doors,max_speed, safety_features,
                  sound_system)
    elif model_code == 4:
       return M04(car_type, color, engine_capacity, tank_capacity,
                  battery_capacity, num_doors,max_speed, safety_features,
                  sound_system)

#-------------------------------------------------------------------------------

class User:
    def __init__(self, username, userregid, usercar):
        self.user_name = username
        self.user_reg_id = userregid
        self.user_car = usercar
    

#-------------------------------------------------------------------------------

def add_user(users_list):
  str1='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
  str2='0123456789'
  reg_1=''
  reg_2=''
  userregid=''
  while True:
    try:
      username = input("Enter Your Good Name: ")
      if not username.isdigit():
        for i in range(3):
          reg_1+=random.choice(str1)
          reg_2+=random.choice(str2)
        userregid = reg_1 + reg_2
        usercar = None 
        user_obj = User(username, userregid, usercar)
        users_list.append(user_obj)
        return user_obj

      else:
        raise ValueError("Invalid Input ! Enter in String.")

    except ValueError as e :
      print("ERROR: ",e)



#-------------------------------------------------------------------------------
def search_the_users(users_list):
  user_id=[]
  for i in users_list:
    user_id.append(i.user_reg_id)

  while True:
    try:
      search_id = get_user_input("Enter Reg. ID: ","Invalid Reg.ID, Correct Enter!",
                             str)
      if search_id in user_id:
        for obj in users_list:
          if search_id == obj.user_reg_id:
            obj_car=obj.user_car
            ser_id_table = PrettyTable()
            ser_id_table .field_names = ["Features", "Value"]
            ser_id_table .align["Features"] = "l"
            ser_id_table .align["Value"] = "l"
            ser_id_table .add_row(["Username", obj.user_name])
            ser_id_table .add_row(["Registration ID", obj.user_reg_id])
            ser_id_table .add_row(["Model Code", obj_car.model_code])
            ser_id_table .add_row(["Model Type", obj_car.model_type])
            ser_id_table .add_row(["Engine Type", obj_car.engine_type])
            ser_id_table.add_row(["Car Type", obj_car.car_type])
            ser_id_table .add_row(["Car Color", obj_car.color])
            ser_id_table .add_row(["Engine Capacity", obj_car.engine_capacity])
            ser_id_table .add_row(["Tank Capacity", f"{obj_car.tank_capacity} Ltr."])
            ser_id_table .add_row(["Battery Capacity", f"{obj_car.battery_capacity} kWh"])
            ser_id_table .add_row(["Number of Doors", obj_car.num_doors])
            ser_id_table .add_row(["Maximum Speed", f"{obj_car.max_speed} km/h"])
            ser_id_table .add_row(["Safety Features",obj_car.safety_features])
            ser_id_table .add_row(["Sound System",obj_car.sound_system])
            print("\nUser Details:")
            print(ser_id_table)
        break

      else:
        raise ValueError("User not found. Enter Valid Reg. ID.")
    except ValueError as e:
      print("ERROR: ",e)
    
      
          
   


#-------------------------------------------------------------------------------
def fetch_all_users(users_list):
    # Creating pretty table for registration IDs and names
    reg_id_table = PrettyTable()
    reg_id_table.field_names = ["Registration ID", "Name"]
    reg_id_table.align["Registration ID"] = "l"
    reg_id_table.align["Name"] = "l"
    for user_obj in users_list:
        reg_id_table.add_row([user_obj.user_reg_id, user_obj.user_name])
    print("\nAll Registered IDs and Names:")
    print(reg_id_table)
#-------------------------------------------------------------------------------
def remove_user(users_list):
  user_id=[]
  for i in users_list:
    user_id.append(i.user_reg_id)

  while True:
    try:
      search_id = get_user_input("Enter Reg. ID: ","Invalid Reg.ID, Correct Enter!",
                             str)
      if search_id in user_id:
        for obj in users_list:
          if search_id == obj.user_reg_id:
            users_list.remove(obj)
            print("Removed Successfully !")
            if os.path.exists(obj.user_name):
              os.remove(obj.user_name)
            else:
              print("The file does not exist")
            break
        break

      else:
        raise ValueError("User not found. Enter Valid Reg. ID.")
    except ValueError as e:
      print("ERROR: ",e)
   
        
  
#-------------------------------------------------------------------------------
def create_file(user_obj_2,user_car_obj):
    with open(user_obj_2.user_name,"w") as f:
        f.write("Car Details\n")
        f.write("-"*40)
        f.write(f'\nUsername : {user_obj_2.user_name}\n')
        f.write(f'Register ID : {user_obj_2.user_reg_id}\n')
        f.write("-"*40)
        f.write(f'\nModel_Code : {user_car_obj.model_code}\n')
        f.write(f'Model_Type : {user_car_obj.model_type}\n')
        f.write(f'Engine_Type : {user_car_obj.engine_type}\n')
        f.write(f'Car_Type : {user_car_obj.car_type}\n')
        f.write(f'Car_Color : {user_car_obj.color}\n')
        f.write(f'Engine_Capacity : {user_car_obj.engine_capacity} \n')
        f.write(f'Tank_Capacity : {user_car_obj.tank_capacity} Ltr.\n')
        f.write(f'Battery_Capacity : {user_car_obj.battery_capacity} kwh\n')
        f.write(f'No_of_Doors :{user_car_obj.num_doors}\n')
        f.write(f'Maximum_Speed : {user_car_obj.max_speed} km/h \n')
        f.write(f'Safety_Feature : {user_car_obj.safety_features}\n')
        f.write(f'Sound_System : {user_car_obj.sound_system}\n')
        f.write("-"*40)
        
        
#-------------------------------------------------------------------------------
def main():
    print("Welcome to Sugam's Automotive Manufacturing Plant\n")
    users_list = []
    all_user=os.listdir()
    
    
    while True:
        try:
          print("\nMenu:")
          print("1. Add User Detail")
          print("2. Fetch Name And Registered ID Of All Users")
          print("3. Fetch Details By ID")
          print("4. Remove Details By ID")
          print("5. Exit")

          choice = input("\nEnter your choice: ")

          if not choice.isdigit():
            raise ValueError("Invalid Input. Enter Integer Value only.")


          choice=int(choice)
          if choice<0 or choice>5:
            raise ValueError("Re-enter input in range 1-5 ")
          else:
              if choice == 1:
                  user_obj_2=add_user(users_list)
              elif choice == 2:
                  if len(users_list) == 0:
                    print("ERROR: User Not Found! Please Register First.")
                    continue
                  else:
                    fetch_all_users(users_list)
                    continue
              elif choice == 3:
                if len(users_list) == 0:
                  print("ERROR: User Not Found! Please Register First.")
                  continue
                else:
                  search_the_users(users_list)
                  continue
              elif choice == 4:
                if len(users_list) == 0:
                  print("ERROR: User Not Found! Please Register First.")
                  continue
                else:
                  remove_user(users_list) 
                  continue
              elif choice == 5:
                  exit()
                  print("-"*20,"Thank you for Visiting.","-"*20)
                  break

        except ValueError as e:
              print("Error : ",e)
        else:
          while True:
            # user_obj_2 --> user name,id,car.
            manufactured_car = manufacture_car(user_obj_2) #car detail of user
            user_obj_2.user_car = manufactured_car
            
            # creating the file
            create_file(user_obj_2,manufactured_car)          

            # Creating pretty table for car details
            car_details_table = PrettyTable()
            car_details_table.field_names = ["Features", "Value"]
            car_details_table.align["Features"] = "l"
            car_details_table.align["Value"] = "l"
            car_details_table.add_row(["Username", user_obj_2.user_name])
            car_details_table.add_row(["Registration ID", user_obj_2.user_reg_id])
            car_details_table.add_row(["Model Code", manufactured_car.model_code])
            car_details_table.add_row(["Model Type", manufactured_car.model_type])
            car_details_table.add_row(["Engine Type", manufactured_car.engine_type])
            car_details_table.add_row(["Car Type", manufactured_car.car_type])
            car_details_table.add_row(["Car Color", manufactured_car.color])
            car_details_table.add_row(["Engine Capacity", manufactured_car.engine_capacity])
            car_details_table.add_row(["Tank Capacity",
                                       f"{manufactured_car.tank_capacity} Ltr."])
            car_details_table.add_row(["Battery Capacity",
                                       f"{manufactured_car.battery_capacity} kWh"])
            car_details_table.add_row(["Number of Doors",
                                       manufactured_car.num_doors])
            car_details_table.add_row(["Maximum Speed",
                                       f"{manufactured_car.max_speed} km/h"])
            car_details_table.add_row(["Safety Features",manufactured_car.safety_features])
            car_details_table.add_row(["Sound System",manufactured_car.sound_system])
            print("\nCar Details:")
            print(car_details_table)
            break
            

if __name__ == "__main__":
    main()
```
